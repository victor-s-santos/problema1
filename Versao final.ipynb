```python
import json, requests

api_historico = 'http://www.mocky.io/v2/598b16861100004905515ec7'
api_clientes = 'http://www.mocky.io/v2/598b16291100004705515ec5'

class Tudo:
    def __init__(self, 
                 api_clientes='http://www.mocky.io/v2/598b16291100004705515ec5', 
                 api_historico='http://www.mocky.io/v2/598b16861100004905515ec7', 
                 cpf_cliente='0000.000.000.01'):
        """Os argumentos nomeados possuem como valor padrão as api's usadas,
        e um cpf determinado"""
        self.api_clientes = api_clientes
        self.api_historico = api_historico
        self.cpf_cliente = cpf_cliente
        
    def conecta_api_historico(self):
        """Este método retorna o conteúdo da api, em uma lista de dicionários"""
        response = requests.get(api_historico)
        conteudo = json.loads(response.content)
        return conteudo
    
    def conecta_api_cliente(self):
        """Este método retorna o conteúdo da api, em uma lista de dicionários"""
        response = requests.get(api_clientes)
        conteudo = json.loads(response.content)
        return conteudo

    def lista_decrescente(self):
        """Retorna uma lista decrescente de acordo com a chave valorTotal"""
        lista_ordenada_decrescente = sorted(self.conecta_api_historico(), 
                                    key=lambda chave: -chave['valorTotal'])
        return lista_ordenada_decrescente
    
    def maior2016(self):
        """Retorna o maior valor total contido na lista_decrescente, assim como 
        o nome do cliente"""
        lista_2016 = []
        self.lista_2016 = lista_2016
        cliente = self.conecta_api_cliente()
        for dados in self.lista_decrescente():
            if dados['data'][-4:] == '2016':
                self.lista_2016.append(dados)
        for i in range(len(cliente)):
            if '0' + cliente[i]['cpf'].replace('-', '.') == self.lista_2016[0]['cliente']:
                return (cliente[i]['nome'], self.lista_2016[0]['valorTotal'])
    
    def fiel(self):
        """Retorna uma tupla, contendo o dicionário['cliente': frequencia] 
        e o cliente mais frequente"""
        nomes = []
        cliente = self.conecta_api_cliente()
        for i in range(len(cliente)):
            nomes.append(cliente[i]['nome'])
        #return(len(nomes))
        nomes_historico = []
        #return(nomes)
        nomes_historico = []
        for nome in range(len(self.lista_decrescente())):
            nomes_historico.append(self.lista_decrescente()[nome]['cliente'])
        nomes_historico[-5] = '0' + nomes_historico[-5]
        #return(nomes_historico)
    
        freq_nomes = []
        for i in range(len(cliente)):
            freq_nomes.append(nomes_historico.count('0' + cliente[i]['cpf']
                                                    .replace('-', '.')))
        nomes_freq = dict(zip(nomes, freq_nomes))     
        return(nomes_freq, max(nomes_freq, key=nomes_freq.get))
    
    def seleciona_vinho(self):
        """É selecionado um vinho para um cliente inserido, 
        de acordo com seu histórico."""
        conteudo = self.conecta_api_historico()
        brinde = []
        for i in range(len(conteudo)):
            if conteudo[i]['cliente'] == self.cpf_cliente:
                brinde.append(conteudo[i]['itens'])

        escolhido = []
        for lista1 in brinde:
            for lista2 in lista1:
                #print(lista2['produto'])
                escolhido.append(lista2['produto'])
        freq_escolhido = []
        for i in range(len(escolhido)):
            freq_escolhido.append(escolhido.count(escolhido[i]))
        produto_freq = dict(zip(escolhido, freq_escolhido))
        return(max(produto_freq, key=produto_freq.get))
    
    

       

        
teste1 = Tudo()
#teste1.conecta_api_historico()
#teste1.conecta_api_cliente()
#teste1.lista_decrescente()
#teste1.maior2016()
#teste1.fiel()
teste1.seleciona_vinho()
```
