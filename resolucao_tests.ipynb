```python
import json, requests

if __name__ == '__main__':
    assert status_code('http://www.mocky.io/v2/598b16291100004705515ec5') == 200
```

```python
def content(api):
    """Deve retornar 200"""
    response = requests.get(api)
    return response.status_code

if __name__ == '__main__':
    assert content('http://www.mocky.io/v2/598b16291100004705515ec5') == 200
    assert content('http://www.mocky.io/v2/598b16861100004905515ec7') == 200
#status code das duas api é 200
```

```python
def conecta_api(api_link):
    """Deve retornar o conteúdo da api"""
    response = requests.get(api_link)
    conteudo = json.loads(response.content)
    return conteudo

a = conecta_api('http://www.mocky.io/v2/598b16861100004905515ec7')[0]
b = conecta_api('http://www.mocky.io/v2/598b16861100004905515ec7')[-1]

if __name__ == '__main__':
    assert type(conecta_api('http://www.mocky.io/v2/598b16861100004905515ec7')) == list
```

```python
if __name__ == '__main__':
    assert a['valorTotal'] > b['valorTotal']
#a lista não está decrescente
```

```python
def lista_clientes_valor_total(api_link):
    """Deve retornar uma lista em ordem decrescente pelo valor total"""
    conteudo = conecta_api(api_link)
    lista_ordenada_decrescente = sorted(conteudo, key=lambda chave: -chave['valorTotal'])
    return lista_ordenada_decrescente

a = lista_clientes_valor_total('http://www.mocky.io/v2/598b16861100004905515ec7')[0]
b = lista_clientes_valor_total('http://www.mocky.io/v2/598b16861100004905515ec7')[-1]

if __name__ == '__main__':
    assert a['valorTotal'] > b['valorTotal']
```

* ## 4 - Recomende um vinho para um determinado cliente a partir do histórico de compras.

```python
api_historico = 'http://www.mocky.io/v2/598b16861100004905515ec7'
api_cliente = 'http://www.mocky.io/v2/598b16291100004705515ec5'
```

```python
def seleciona_vinho(api_historico,cpf_cliente):
    conteudo = conecta_api(api_historico)
    brinde = []
    for i in range(len(conteudo)):
        if conteudo[i]['cliente'] == cpf_cliente:
            brinde.append(conteudo[i]['itens'])

    escolhido = []
    for lista1 in brinde:
        for lista2 in lista1:
            #print(lista2['produto'])
            escolhido.append(lista2['produto'])
    freq_escolhido = []
    for i in range(len(escolhido)):
        freq_escolhido.append(escolhido.count(escolhido[i]))
    produto_freq = dict(zip(escolhido, freq_escolhido))
    return(max(produto_freq, key=produto_freq.get))

seleciona_vinho('http://www.mocky.io/v2/598b16861100004905515ec7', '0000.000.000.02')
```
