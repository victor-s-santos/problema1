```python
import json, requests
api_historico = 'http://www.mocky.io/v2/598b16861100004905515ec7'
api_cliente = 'http://www.mocky.io/v2/598b16291100004705515ec5'

def conecta_api(api_link):
    """Deve retornar o conteúdo da api"""
    response = requests.get(api_link)
    conteudo = json.loads(response.content)
    return conteudo
```

## ordenando os clientes pelo valorTotal em forma decrescente


```python
lista_ordenada_decrescente = sorted(conecta_api(api_historico), 
                                    key=lambda chave: -chave['valorTotal'])

lista_ordenada_decrescente
```

## pegando todos os clientes do ano de 2016

```python
lista_2016 = []
for dados in lista_ordenada_decrescente:
    if dados['data'][-4:] == '2016':
        lista_2016.append(dados)
```

## maior valor total de 2016, e o cpf do cliente, pois esta lista foi obtida da lista ordenada decrescente

```python
lista_2016[0]['valorTotal']
```

## lista dos nomes dos clientes(na api dos clientes):

```python
nomes = []
for i in range(len(cliente)):
    nomes.append(cliente[i]['nome'])
len(nomes)
```

## lista dos nomes e quantidade de vezes (na api de histórico)

```python
nomes_historico = []
for nome in range(len(lista_ordenada_decrescente)):
    nomes_historico.append(lista_ordenada_decrescente[nome]['cliente'])
nomes_historico
```

```python
#checando se bate a quantidade de registros
len(nomes_historico) == 50
```

## consertando o objeto de índice i = -5

```python
#como é apenas um registro, farei a inserção manual
nomes_historico[-5] = '0' + nomes_historico[-5]
nomes_historico
```

## fazendo uma lista das frequências

```python
freq_nomes = []
for i in range(len(cliente)):
    freq_nomes.append(nomes_historico.count('0' + cliente[i]['cpf'].replace('-', '.')))
freq_nomes
```

## testando se o número total é igual a 50

```python
sum(freq_nomes) == 50
```

## testando se o número de valores é 10

```python
len(freq_nomes) == 10
```

## Como a ordem da lista de frequencias segue a ordem da lista nomes

```python
nomes
freq_nomes
nomes_freq = dict(zip(nomes, freq_nomes))
nomes_freq
```

## Com este dicionário de fidelidade, posso retornar o mais e o menos fiel, respectivamente

```python
max(nomes_freq, key=nomes_freq.get), min(nomes_freq, key=nomes_freq.get)
```

## Recomendando um vinho de acordo com o histórico do cliente

```python
def seleciona_vinho(api_historico,cpf_cliente):
    conteudo = conecta_api(api_historico)
    brinde = []
    for i in range(len(conteudo)):
        if conteudo[i]['cliente'] == cpf_cliente:
            brinde.append(conteudo[i]['itens'])

    escolhido = []
    for lista1 in brinde:
        for lista2 in lista1:
            #print(lista2['produto'])
            escolhido.append(lista2['produto'])
    freq_escolhido = []
    for i in range(len(escolhido)):
        freq_escolhido.append(escolhido.count(escolhido[i]))
    produto_freq = dict(zip(escolhido, freq_escolhido))
    return(max(produto_freq, key=produto_freq.get))

seleciona_vinho('http://www.mocky.io/v2/598b16861100004905515ec7', '0000.000.000.02')
```
